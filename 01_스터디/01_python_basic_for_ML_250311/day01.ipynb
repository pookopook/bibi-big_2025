### 데이터란?
 0. 데이터의 정의
    - 데이터(Data)란 현실 세계에서 관찰되거나 측정된 사실, 개념, 명령 등의 값을 의미합니다.
    - DIKW 피라미드: 데이터(Data) → 정보(Information) → 지식(Knowledge) → 지혜(Wisdom)
 1. 일상(업무)에서 겪는 문제, 왜 분석이 안되는가?
    - 엑셀로 데이터를 분석하는 방법(vlookup, pivot)
      1. vlookup - 두 테이블의 연결 -> 새로운 정보 생산
      2. pivot - 데이터의 집계(**그룹**별, 합, 평균)
    - 올바른 형태로 데이터를 만드는 방법 (DB 정규화)
      1. 크로스탭 쿼리 정규화하는 간단한 예시
 2. 엑셀에서 파이썬으로 넘어가기, why?
    - 대량의 데이터(빅데이터)
    - 여러 테이블의 연계
    - 분석 ex. A/B test
    - 예측 ml

#### 라이브러리 불러오기
 - **라이브러리(Library)** 는 특정 기능을 수행하는 함수, 클래스, 모듈들의 모음입니다.
   - 공구상자 같은 것으로 생각하면 됩니다.
   - 머신러닝을 위한 공구상자, 통계분석을 위한 공구상자, 데이터 크롤링을 위한 공구상자 등...

 - 파이썬에는 두가지 핵심 라이브러리가 있습니다.
    > ``` python

    > ```

  - 마찬가지로 많은 라이브러리가 전통적으로 사용하는 요약어가 있습니다.
  - 많은 라이브러리들은 자체 홈페이지가 있고, 거기서 제시하는 방식을 사용합니다.
